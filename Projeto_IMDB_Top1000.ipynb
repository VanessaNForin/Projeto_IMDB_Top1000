###**Qual per√≠odo de anos possui as maiores notas e maiores bilheterias? Existiu um "per√≠odo de ouro" nessa lista?**

```python
#Descobrindo o ano mais antigo e mais recente da coluna Released_Year
#Para isso, terei que fazer um tratamento na coluna, pois um dos valores √© PG, o que impossibilita a analise que deve ser feita apenas ocm n√∫mero
#Eis a linha com o valor PG na Released_Year

df_imdb_top[df_imdb_top['Released_Year'] == 'PG']
```

```python
#Fazendo a substitui√ß√£o
#De acordo com o pr√≥prio site do IMDB, o filme foi lan√ßado em 1995

df_imdb_top.loc[968, 'Released_Year'] = '1995'
```

```python
#Vamos fazer a convers√£o da coluna agora

df_imdb_top['Released_Year'] = df_imdb_top['Released_Year'].astype(int)
```

```python
#Encontrando o ano mais antigo e recente

ano_mais_recente = df_imdb_top['Released_Year'].max()
ano_mais_antigo = df_imdb_top['Released_Year'].min()

print(f'O ano mais antigo √© : {ano_mais_antigo} e o mais recente √©: {ano_mais_recente}')
```

```python
#Com isso, posso definir um intervalor de anos
#Para n√£o ficar muito longo, irei fazer intervalos de 20 anos.

#1920 - 1939
#1940 - 1959
#1960 - 1979
#1980 - 1999
#2000 - 2020

faixa_de_anos = [1920, 1940, 1960, 1980, 2000, 2020]

df_imdb_top['Faixa_Anos'] = pd.cut(df_imdb_top['Released_Year'], bins = faixa_de_anos, labels = ['1920-1939', '1940-1959', '1960-1979', '1980-1999', '2000-2020'])
```

```python
#Convertendo a coluna Gross para float

df_imdb_top["Gross"] = pd.to_numeric(df_imdb_top["Gross"].str.replace(",", "").str.strip(), errors="coerce").fillna(0)
```

```python
#Criando os filtros da coluna Gross com a separa√ß√£o dos anos e evitando os zeros

filmes_1920_1939 = df_imdb_top[(df_imdb_top['Faixa_Anos'] == '1920-1939') & (df_imdb_top['Gross'] >  0)]
filmes_1940_1959 = df_imdb_top[(df_imdb_top['Faixa_Anos'] == '1940-1959') & (df_imdb_top['Gross'] >  0)]
filmes_1960_1979 = df_imdb_top[(df_imdb_top['Faixa_Anos'] == '1960-1979') & (df_imdb_top['Gross'] >  0)]
filmes_1980_1999 = df_imdb_top[(df_imdb_top['Faixa_Anos'] == '1980-1999') & (df_imdb_top['Gross'] >  0)]
filmes_2000_2020 = df_imdb_top[(df_imdb_top['Faixa_Anos'] == '2000-2020') & (df_imdb_top['Gross'] >  0)]
```

```python
#Criando as somas e m√©dias

soma_gross_1920_1939 = filmes_1920_1939['Gross'].sum()
media_gross_1920_1939 = filmes_1920_1939['Gross'].mean()
soma_gross_1940_1959 = filmes_1940_1959['Gross'].sum()
media_gross_1940_1959 = filmes_1940_1959['Gross'].mean()
soma_gross_1960_1979 = filmes_1960_1979['Gross'].sum()
media_gross_1960_1979 = filmes_1960_1979['Gross'].mean()
soma_gross_1980_1999 = filmes_1980_1999['Gross'].sum()
media_gross_1980_1999 = filmes_1980_1999['Gross'].mean()
soma_gross_2000_2020 = filmes_2000_2020['Gross'].sum()
media_gross_2000_2020 = filmes_2000_2020['Gross'].mean()
```

```python
#Criando um df com as m√©dias e somas

data_frame_bilheteria = [
    ('1920-1939', soma_gross_1920_1939, media_gross_1920_1939),
    ('1940-1959', soma_gross_1940_1959, media_gross_1940_1959),
    ('1960-1979', soma_gross_1960_1979, media_gross_1960_1979),
    ('1980-1999', soma_gross_1980_1999, media_gross_1980_1999),
    ('2000-2020', soma_gross_2000_2020, media_gross_2000_2020)
    ]

df_bilheteria = pd.DataFrame(data_frame_bilheteria, columns=['Intervalo_Anos', 'Somas', 'M√©dias'])
```

###**Qual o g√™nero mais popular?**

```python
df_imdb_top
```

```python
#Transformando a coluna Genre em uma lista, garantindo que todos os elementos estejam separados

generos_lista = df_imdb_top['Genre'].str.split(', ').explode().tolist()
```

```python
generos_lista
```

```python
#Usando o Counter para encontrar os g√™neros mais popular e calcular a porcentagem

from collections import Counter

genero_mais_popular = Counter(generos_lista).most_common(1)[0][0]
qnt_vezes = Counter(generos_lista).most_common(1)[0][1]

linhas, colunas = df_imdb_top.shape
porcentagem_gen_popular = round((qnt_vezes/linhas)* 100, 2)
```

**<h3>Resultado<h3>**

```python
print(f'O g√™nero mais popular √© o {genero_mais_popular} e aparece {qnt_vezes} vezes')
print(f'Ele equivale a {porcentagem_gen_popular}% da lista')
```

###**Quais s√£o os atores que mais aparecem?**

```python
#Convertendo as colunas de atores em uma lista

lista1 = df_imdb_top['Star1'].values.tolist()
lista2 = df_imdb_top['Star2'].values.tolist()
lista3 = df_imdb_top['Star3'].values.tolist()
lista4 = df_imdb_top['Star4'].values.tolist()

lista_atores = lista1 + lista2 + lista3 + lista4
```

```python
#Encontrando os atores mais populares e os colocando em um top 10

ator_popular = Counter(lista_atores).most_common(10)

nomes_atores = [item[0] for item in ator_popular]
qnt_vezes_ator = [item[1] for item in ator_popular]

posicao = 1
print('Os 10 atores mais populares:')
for nomes,qnt in zip(nomes_atores,qnt_vezes_ator):
  print(f'{posicao}¬∫ : {nomes}, {qnt}')
  posicao += 1
```

**<h3>Resultado<h3>**

Destaca-se a presente de massiva maioria de atores de filmes norte-americanos. Abaixo, √© poss√≠vel ver que a presen√ßa de atrizes ocorre apenas em 15¬∫, sendo essa a Scarlett Johansson

```python
#Lista para mostrar atrizes

Counter(lista_atores).most_common(20)
```
